```python
# 2.
import collections

def most_elements(language):
    """ Return a list of lines after inserting a 
word in a specific line. """
    count_time = collections.Counter(language)
    #count_time.sort()
    most_time = count_time.most_common(2)
    most_element = most_time[0][0]
    print(most_time)
    print(most_element)
    
def main():
    language = ['PHP', 'PHP', 'Python', 'PHP', 'Python', 'JS', 'Python', 'Python','PHP', 'Python']
    most_elements(language)
    
if __name__ == "__main__":
    main()
```
